## Train on CIFAR-10

```python
def load_data(train_batch_size, test_batch_size):
    train_loader = torch.utils.data.DataLoader(
        datasets.CIFAR10('data', train=True, download=True,
                         transform=transforms.Compose([
                             transforms.Resize((32, 32)),
                             transforms.ToTensor(),
                             transforms.Normalize((0.4914, 0.4822, 0.4465), (0.247, 0.243, 0.261))
                         ])),
        batch_size=train_batch_size, shuffle=True)

    test_loader = torch.utils.data.DataLoader(
        datasets.CIFAR10('data', train=False, transform=transforms.Compose([
            transforms.Resize((32, 32)),
            transforms.ToTensor(),
            transforms.Normalize((0.4914, 0.4822, 0.4465), (0.247, 0.243, 0.261))
        ])),
        batch_size=test_batch_size, shuffle=False)

    return (train_loader, test_loader)

train_batch_size = 64
test_batch_size = 1000
train_loader, test_loader = load_data(train_batch_size, test_batch_size)
```

```python
def train(model, optimizer, epoch, train_loader, log_interval):
    model.train()
    loss_fn = torch.nn.CrossEntropyLoss()
    num_correct = 0
    train_loss = 0
    count = 0
    num_samples = 0

    for batch_idx, (data, target) in enumerate(train_loader):
        count += 1
        data, target = Variable(data), Variable(target)
        optimizer.zero_grad()

        output = model(data)
        loss = loss_fn(output, target)
        loss.backward()
        optimizer.step()

        train_loss += loss.data

        predict = np.argmax(output.data, axis=1)
        num_samples += target.size(0)
        num_correct += (predict == target).sum().item()
        acc = 100.0 * num_correct / num_samples

        if batch_idx % log_interval == 0:
            print('Training, Epoch {} [{}/{}]\tLoss: {:.4f}'.format(
                epoch, batch_idx * len(data), len(train_loader.dataset),
                loss.data))
            print("Accuracy: ", round(acc, 2))
    train_loss /= count
    return train_loss, round(acc, 2)
```

```python
def test(model, epoch, test_loader):
    model.eval()
    test_loss = 0
    num_correct = 0
    
    loss_fn = torch.nn.CrossEntropyLoss(size_average=False)

    for data, target in test_loader:
        data, target = Variable(data), Variable(target)
        
        output = model(data)
        test_loss += loss_fn(output, target).data

        predict = np.argmax(output.data, axis=1)
        num_correct = num_correct + np.equal(predict, target.data).sum()

    test_loss /= len(test_loader.dataset)
    print('\nTesting, Epoch {} , Avg. loss: {:.4f}, Accuracy: {}/{} ({:.0f}%)\n'.format(epoch,
        test_loss, num_correct, len(test_loader.dataset),
        100. * num_correct / len(test_loader.dataset)))
```

#### CNN 

```python
class CNN(nn.Module):
    def __init__(self, hidden1_dim):
        super(CNN, self).__init__()
        self.conv1 = nn.Conv2d(3, hidden1_dim, 5)  
        self.conv2 = nn.Conv2d(hidden1_dim, 32, 5)
        
        self._to_linear = None
        self._get_conv_output_size()

        self.fc1 = nn.Linear(self._to_linear, 256)
        self.fc2 = nn.Linear(256, 10)

    def _get_conv_output_size(self):
        # compute the size of the output after conv layers
        dummy_input = torch.randn(1, 3, 32, 32)  
        x = F.relu(self.conv1(dummy_input))
        x = F.relu(F.max_pool2d(self.conv2(x), 2))
        self._to_linear = x.view(1, -1).size(1)  
        
    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = F.relu(F.max_pool2d(self.conv2(x), 2))
        x = x.view(x.size(0), -1)  # Flatten the tensor
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return F.log_softmax(x, dim=1)
```

```python
filter_sizes = [16, 44, 80]# 60, 80, 100]

# dictionary of models with different filter sizes
models = {f'CNN_hidden1_dim{size}': CNN(hidden1_dim=size) for size in filter_sizes}

optimizers = {name: optim.SGD(model.parameters(), lr=0.01, momentum=0.5) for name, model in models.items()}

train_losses = {name: [] for name in models}
train_accuracies = {name: [] for name in models}

epochs = 10
log_interval = 200

# training
for model_name, model in models.items():
    print(f"Training {model_name}")
    optimizer = optimizers[model_name]
    for epoch in range(1, epochs + 1):
        loss, acc = train(model, optimizer, epoch, train_loader, log_interval)
        train_losses[model_name].append(loss)
        train_accuracies[model_name].append(acc)
        test(model, epoch, test_loader)
```

```python
plt.figure(figsize=(10,5))
for model_name, losses in train_losses.items():
    plt.plot(range(1, epochs + 1), losses, label=model_name)
plt.title('Training Loss of All CNN Models with Different Hidden Dimensions')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python

plt.figure(figsize=(10,5))
for model_name, accs in train_accuracies.items():
    plt.plot(range(1, epochs + 1), accs, label=model_name)
plt.title('Training Accuracy of All CNN Models with Different Hidden Dimensions')
plt.xlabel('Epochs')
plt.ylabel('Accuracy (%)')
plt.legend()
plt.show()
```

### DNN

```python
class DNN(nn.Module):
    def __init__(self, input_size, hidden1_dim):
        super(DNN, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden1_dim)  # First hidden layer
        self.fc2 = nn.Linear(hidden1_dim, 256)         # Second hidden layer
        self.fc3 = nn.Linear(256, 10)                  # Output layer

    def forward(self, x):
        x = x.view(x.size(0), -1)  # Flatten the input tensor
        x = F.relu(self.fc1(x))    # Apply ReLU activation after the first layer
        x = F.relu(self.fc2(x))    # Apply ReLU activation after the second layer
        x = self.fc3(x)            # Output layer (no activation for logits)
        return F.log_softmax(x, dim=1)
```

```python
hidden_sizes = [128, 200, 260]  

models = {f'DNN_hidden1_dim{size}': DNN(input_size=3*32*32, hidden1_dim=size) for size in hidden_sizes}

optimizers = {name: optim.SGD(model.parameters(), lr=0.01, momentum=0.5) for name, model in models.items()}

train_losses = {name: [] for name in models}
train_accuracies = {name: [] for name in models}

epochs = 10
log_interval = 200

for model_name, model in models.items():
    print(f"Training {model_name}")
    optimizer = optimizers[model_name]
    for epoch in range(1, epochs + 1):
        loss, acc = train(model, optimizer, epoch, train_loader, log_interval)
        train_losses[model_name].append(loss)
        train_accuracies[model_name].append(acc)
        test(model, epoch, test_loader)
```

```python
plt.figure(figsize=(10,5))
for model_name, losses in train_losses.items():
    plt.plot(range(1, epochs + 1), losses, label=model_name)
plt.title('Training Loss of All DNN Models')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
for model_name, accs in train_accuracies.items():
    plt.plot(range(1, epochs + 1), accs, label=model_name)
plt.title('Training Accuracy of All DNN Models')
plt.xlabel('Epochs')
plt.ylabel('Accuracy (%)')
plt.legend()
plt.show()
```
